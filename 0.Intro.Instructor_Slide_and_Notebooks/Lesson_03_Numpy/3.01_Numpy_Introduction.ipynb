# 📚 **Part 1: Fundamentals of NumPy**

---

## 🤔 **What is NumPy?**

**NumPy** stands for **Numerical Python**.

It's a **free and open-source Python library** that provides:
- 🔢 **Powerful array objects** for storing data
- ⚡ **Fast mathematical operations** on arrays
- 🛠️ **Useful functions** for scientific computing
- 📊 **Foundation for data science** and machine learning

---

## 🎯 **Core Features of NumPy**

### **1. Array-Based Computing**
- NumPy provides the `ndarray` (n-dimensional array) object
- Similar to Python lists but **much more powerful**
- Can store millions of numbers efficiently

### **2. Mathematical Operations**
- Perform calculations on entire arrays at once
- No need to loop through elements manually
- Supports linear algebra, statistics, and more

### **3. Memory Efficiency**
- Arrays are stored in **continuous memory locations**
- Much faster access than Python lists
- Critical for large datasets

---

## 💡 **Real-World Examples**

NumPy is used everywhere in data science:

| Application | How NumPy Helps |
|-------------|----------------|
| 🏠 **Real Estate** | Store and analyze thousands of house prices |
| 📈 **Stock Market** | Process millions of stock prices over time |
| 🎬 **Image Processing** | Store pixel values (images are just arrays!) |
| 🧬 **Genomics** | Analyze DNA sequences (arrays of genetic data) |
| 🤖 **Machine Learning** | Store training data and model parameters |
| 🌡️ **Weather Forecasting** | Process temperature, pressure, humidity arrays |

---

## 📝 **Important Note**

**The array object in NumPy is called `ndarray`** (n-dimensional array).

Think of it as:

- **1D array** = A list of numbers (vector)

## ⚡ **1.1 Why NumPy? (Advantages Over Python Lists)**

You might wonder: *"Python already has lists. Why do I need NumPy?"*

Great question! Let's see why NumPy arrays are **WAY better** for data science:

---

### **🚀 Advantage #1: Speed (10-100x Faster!)**

**Python List:**
```python
# Slow - Loops through each element
result = []
for i in range(1000000):
    result.append(i * 2)
```

**NumPy Array:**
```python
# Fast - Vectorized operation (no loop needed!)
result = np.arange(1000000) * 2
```

**Why faster?**
- NumPy is written in C (compiled language)
- Operations happen in parallel on the CPU
- No Python overhead for each element

---

### **💾 Advantage #2: Memory Efficiency**

**Python List:**
- Each element is a separate Python object
- Scattered across different memory locations
- Takes more space

**NumPy Array:**
- All elements stored **continuously in memory**
- More compact storage
- Faster access (CPU cache-friendly)

**Example:**
- Python list of 1 million integers: ~28 MB
- NumPy array of 1 million integers: ~4 MB
- **7x less memory!**

---

### **📊 Advantage #3: Convenient Operations**

**Python List:**
```python
# Manual element-wise operations
list1 = [1, 2, 3, 4]
list2 = [5, 6, 7, 8]
result = [a + b for a, b in zip(list1, list2)]
```

**NumPy Array:**
```python
# Simple and readable
arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([5, 6, 7, 8])
result = arr1 + arr2  # That's it!
```

---

### **🛠️ Advantage #4: Built-in Functions**

NumPy provides hundreds of useful functions:

```python
arr = np.array([1, 2, 3, 4, 5])

# Statistics
print(arr.mean())    # Average
print(arr.std())     # Standard deviation
print(arr.sum())     # Sum
print(arr.max())     # Maximum

# Linear algebra
print(arr.dot(arr))  # Dot product
# And many more!
```

---

### **📈 Advantage #5: Foundation for Data Science**

**All major libraries use NumPy:**
- 🐼 **Pandas**: Built on top of NumPy arrays
- 📊 **Matplotlib**: Uses NumPy for plotting
- 🤖 **Scikit-learn**: All data is NumPy arrays
- 🧠 **TensorFlow/PyTorch**: Internal tensors are NumPy-like

**Learn NumPy = Learn the foundation of everything!**

---

### **📊 Quick Comparison Table**

| Feature | Python List | NumPy Array |
|---------|-------------|-------------|
| **Speed** | Slow (Python loops) | ⚡ Fast (C implementation) |
| **Memory** | More (separate objects) | 💾 Less (continuous storage) |
| **Operations** | Manual loops needed | ✨ Vectorized (automatic) |
| **Functions** | Basic (len, sum, etc.) | 🛠️ Hundreds (mean, std, etc.) |

**When to use NumPy Arrays:**

### **💡 Common Import Patterns**

**✅ Correct (Standard):**
```python

```

```python

```

**⚠️ Also Avoid:**
```python

```


# 📦 **Part 2: NumPy Array Object (ndarray)**

---

## 🎯 **What is an ndarray?**

**`ndarray`** = **n-dimensional array** (the core of NumPy!)

Think of it as a container that can hold:
- Numbers arranged in **1 dimension** (like a list)
- Numbers arranged in **2 dimensions** (like a table/spreadsheet)
- Numbers arranged in **3+ dimensions** (like a cube or higher)

---

## 🔨 **Creating Your First NumPy Array**

The most common way to create a NumPy array is using the `array()` function.

**Syntax:**
```python
np.array(data)
```

Let's create a simple array:


````python
# =============================================================================
# EXAMPLE: CREATE A BASIC 1D NUMPY ARRAY
# =============================================================================
# Purpose: Demonstrate how to create a NumPy array and verify its type

# Import NumPy with the standard alias
import numpy as np

# Create a 1D array from a Python list
# - Pass a list [10, 20, 30, 40, 50] to np.array()
# - NumPy converts it to an ndarray object
arr = np.array([10, 20, 30, 40, 50])

# Display the array
print("📊 Array contents:")
print(arr)
print()

# Check the type - should be numpy.ndarray
print("🔍 Array type:")
print(type(arr))
print()

# Additional information
print("📏 Array shape:", arr.shape)  # (5,) means 5 elements in 1 dimension
print("📦 Array size:", arr.size)    # Total number of elements
print("🔢 Array dtype:", arr.dtype)  # Data type (int64, float64, etc.)
````

---

## 🧐 **Understanding the Output**

**What we see:**
```
📊 Array contents:
[10 20 30 40 50]

🔍 Array type:
<class 'numpy.ndarray'>

📏 Array shape: (5,)
📦 Array size: 5
🔢 Array dtype: int64
```

**What this means:**

1. **Array contents:** `[10 20 30 40 50]`
   - Notice: No commas between elements (NumPy display format)
   - All elements stored in a single row

2. **Type:** `numpy.ndarray`
   - Confirms we created a NumPy array
   - Different from Python list!

3. **Shape:** `(5,)`
   - 5 elements in dimension 0
   - The comma indicates it's a 1D array

4. **Size:** `5`
   - Total number of elements

5. **dtype:** `int64`
   - All elements are 64-bit integers
   - NumPy inferred this from the input

---

## 💡 **Python List vs NumPy Array**

| Aspect | Python List | NumPy Array |
|--------|-------------|-------------|
| **Creation** | `[10, 20, 30]` | `np.array([10, 20, 30])` |
| **Display** | `[10, 20, 30]` | `[10 20 30]` (no commas) |
| **Type** | `<class 'list'>` | `<class 'numpy.ndarray'>` |
| **Operations** | Element-by-element loops | Vectorized (fast!) |
| **Memory** | More | Less (continuous) |

## 🌐 **2.1 Creating Multi-Dimensional Arrays**

NumPy arrays can have **any number of dimensions**. Let's explore the most common ones:

- **0D Array** = Scalar (a single number)
- **1D Array** = Vector (a list of numbers)
- **2D Array** = Matrix (a table of numbers)
- **3D Array** = Tensor (a cube of numbers)

---

### **Understanding Dimensions**

Think of dimensions like this:

| Dimensions | Real-World Example | Visual Representation |
|------------|-------------------|----------------------|
| **0D** | A single temperature reading | `42` |
| **1D** | Daily temperatures for a week | `[20, 22, 19, 23, 25, 21, 20]` |
| **2D** | Temperatures for multiple cities over a week | A table (rows = cities, cols = days) |
| **3D** | Temperatures for multiple cities, multiple weeks, multiple times | A cube of data |

---

### **Let's Create Each Type!**


````python
# =============================================================================
# CREATE MULTI-DIMENSIONAL ARRAYS
# =============================================================================
# Purpose: Demonstrate 0D, 1D, 2D, and 3D array creation

import numpy as np

print("=" * 70)
print("CREATING MULTI-DIMENSIONAL ARRAYS")
print("=" * 70)
print()

# -----------------------------------------------------------------------------
# 0D ARRAY (SCALAR) - A single number
# -----------------------------------------------------------------------------
# Purpose: Store a single value
# Example: A person's age, a single temperature reading

arr0 = np.array(24)

print("📍 0D Array (Scalar):")
print("   Value:", arr0)
print("   Shape:", arr0.shape)  # () means 0 dimensions
print("   Dimensions:", arr0.ndim)  # 0
print("   Real-world use: A single value like age, score, temperature")
print()

# -----------------------------------------------------------------------------
# 1D ARRAY (VECTOR) - A list of numbers
# -----------------------------------------------------------------------------
# Purpose: Store a sequence of values
# Example: Daily stock prices, student test scores

arr1 = np.array([1, 2, 3, 4])

print("➡️ 1D Array (Vector):")
print("   Value:", arr1)
print("   Shape:", arr1.shape)  # (4,) means 4 elements in 1 dimension
print("   Dimensions:", arr1.ndim)  # 1
print("   Real-world use: Time series data, list of prices, grades")
print()

# -----------------------------------------------------------------------------
# 2D ARRAY (MATRIX) - A table of numbers
# -----------------------------------------------------------------------------
# Purpose: Store data in rows and columns
# Example: Spreadsheet, grayscale image, dataset with features

arr2 = np.array([[1, 1, 1],    # Row 1
                 [1, 2, 1]])   # Row 2

print("📊 2D Array (Matrix):")
print("   Value:")
print(arr2)
print("   Shape:", arr2.shape)  # (2, 3) means 2 rows, 3 columns
print("   Dimensions:", arr2.ndim)  # 2
print("   Real-world use: Excel sheet, CSV data, grayscale image")
print()

# -----------------------------------------------------------------------------
# 3D ARRAY (TENSOR) - A cube of numbers
# -----------------------------------------------------------------------------
# Purpose: Store data in 3 dimensions
# Example: Color images (height × width × color channels), video frames

arr3 = np.array([[[1, 1, 1],    # Matrix 1, Row 1
                  [2, 2, 2]],   # Matrix 1, Row 2

                 [[3, 3, 3],    # Matrix 2, Row 1
                  [4, 4, 4]]])  # Matrix 2, Row 2

print("🧊 3D Array (Tensor):")
print("   Value:")
print(arr3)
print("   Shape:", arr3.shape)  # (2, 2, 3) means 2 matrices, 2 rows each, 3 columns each
print("   Dimensions:", arr3.ndim)  # 3
print("   Real-world use: Color images (RGB), video data, 3D models")
print()

print("=" * 70)
````

---

## 🔍 **Understanding the Structure**

### **Visual Breakdown:**

**0D Array:** Just a number
```
24
```

**1D Array:** A row of numbers
```
[1, 2, 3, 4]
```

**2D Array:** A table (rows × columns)
```
[[1, 1, 1],
 [1, 2, 1]]
```
- 2 rows
- 3 columns
- Shape: (2, 3)

**3D Array:** Multiple tables stacked
```
[[[1, 1, 1],
  [2, 2, 2]],      ← First matrix

 [[3, 3, 3],
  [4, 4, 4]]]      ← Second matrix
```
- 2 matrices
- Each matrix has 2 rows
- Each row has 3 columns
- Shape: (2, 2, 3)

---

## 📏 **Understanding Shape**

The `.shape` attribute tells us the size in each dimension:

| Array Type | Shape | Meaning |
|------------|-------|---------|
| `arr0` | `()` | 0 dimensions (scalar) |
| `arr1` | `(4,)` | 4 elements in dimension 0 |
| `arr2` | `(2, 3)` | 2 rows, 3 columns |
| `arr3` | `(2, 2, 3)` | 2 matrices, 2 rows each, 3 columns each |

**Rule:** Shape has as many numbers as dimensions!

## 💡 **Key Observations**

### **1. Nested Structure:**
- **0D arrays** are individual numbers
- **1D arrays** have 0D arrays as elements (numbers in a list)
- **2D arrays** have 1D arrays as elements (rows in a table)
- **3D arrays** have 2D arrays as elements (matrices in a cube)

### **2. Dimensionality Increases:**
- Each dimension adds a level of nesting
- More brackets `[` = higher dimension

### **3. Shape Interpretation:**
- **1st number** = size of outermost dimension
- **2nd number** = size of next dimension
- And so on...
