# ➕ **Part 1: Arithmetic Operations**

---

## 🤔 **What Are Element-Wise Operations?**

**Element-wise** means: perform the operation on corresponding elements.

**Example:**
```
[1, 2, 3] + [10, 20, 30] = [11, 22, 33]
 ↓  ↓  ↓     ↓   ↓   ↓      ↓   ↓   ↓
 1+10  2+20  3+30
```

**Why powerful?**
- ✅ No loops needed!
- ✅ Much faster than Python lists
- ✅ Clean, readable code

---

## ➕ **1.1 Addition - np.add()**

**Purpose:** Add corresponding elements from two arrays

**Syntax:** `np.add(array1, array2)` or `array1 + array2`

---


````python
# =============================================================================
# EXAMPLE 1: ADDITION
# =============================================================================
# Purpose: Demonstrate element-wise addition

import numpy as np

# Create two 1D arrays
a = np.array([30, 20, 10])
b = np.array([10, 20, 30])

print("=" * 70)
print("ADDITION: ELEMENT-WISE")
print("=" * 70)
print()

print("📊 Array a:", a)
print("📊 Array b:", b)
print()

# Method 1: Using np.add()
result = np.add(a, b)
print("➕ Result (using np.add):", result)
print()

# Method 2: Using + operator (more common!)
result2 = a + b
print("➕ Result (using +):", result2)
print()

print("💡 What happened:")
print("   [30, 20, 10] + [10, 20, 30]")
print("    ↓   ↓   ↓      ↓   ↓   ↓")
print("   [40, 40, 40]")
print()

# Real-world example
print("🌍 Real-world Example:")
print("   Jan sales: ", a, "→ [$30K, $20K, $10K]")
print("   Feb sales: ", b, "→ [$10K, $20K, $30K]")
print("   Total:     ", result, "→ [$40K, $40K, $40K]")
print()

print("=" * 70)
````

---

## ➖ **1.2 Subtraction - np.subtract()**

**Purpose:** Subtract corresponding elements

**Syntax:** `np.subtract(array1, array2)` or `array1 - array2`

---


````python
# =============================================================================
# EXAMPLE 2: SUBTRACTION
# =============================================================================
# Purpose: Demonstrate element-wise subtraction

import numpy as np

# Create two 2D arrays (2 rows × 3 columns each)
a = np.array([[30, 40, 60],
              [50, 70, 90]])

b = np.array([[10, 20, 30],
              [40, 30, 80]])

print("=" * 70)
print("SUBTRACTION: ELEMENT-WISE")
print("=" * 70)
print()

print("📊 Array a:")
print(a)
print()

print("📊 Array b:")
print(b)
print()

# Perform subtraction
result = np.subtract(a, b)
# Or: result = a - b

print("➖ Result (a - b):")
print(result)
print()

print("💡 What happened:")
print("   Row 1: [30-10, 40-20, 60-30] = [20, 20, 30]")
print("   Row 2: [50-40, 70-30, 90-80] = [10, 40, 10]")
print()

# Real-world example
print("🌍 Real-world Example:")
print("   Revenue:  $", a[0])
print("   Costs:    $", b[0])
print("   Profit:   $", result[0])
print()

print("=" * 70)
````

---

## ✖️ **1.3 Multiplication - np.multiply()**

**Purpose:** Multiply corresponding elements

**Syntax:** `np.multiply(array1, array2)` or `array1 * array2`

**Note:** This is **element-wise**, NOT matrix multiplication!

---


````python
# =============================================================================
# EXAMPLE 3: MULTIPLICATION
# =============================================================================
# Purpose: Demonstrate element-wise multiplication

import numpy as np

# Create two 1D arrays
a = np.array([30, 20, 10])
b = np.array([10, 20, 30])

print("=" * 70)
print("MULTIPLICATION: ELEMENT-WISE")
print("=" * 70)
print()

print("📊 Array a:", a)
print("📊 Array b:", b)
print()

# Perform multiplication
result = np.multiply(a, b)
# Or: result = a * b

print("✖️ Result (a * b):", result)
print()

print("💡 What happened:")
print("   [30, 20, 10] * [10, 20, 30]")
print("    ↓   ↓   ↓      ↓   ↓   ↓")
print("   [300, 400, 300]")
print()

# Scalar multiplication
print("🔢 Bonus: Scalar Multiplication")
scalar = 2
print("   ", a, "* 2 =", a * scalar)
print("   (Multiplies each element by 2)")
print()

# Real-world example
prices = np.array([10, 20, 30])  # Price per unit
quantities = np.array([5, 3, 2])  # Quantity sold
revenue = prices * quantities
print("🌍 Real-world Example:")
print("   Prices:     $", prices)
print("   Quantities:  ", quantities)
print("   Revenue:    $", revenue, "→ Total: $", revenue.sum())
print()

print("=" * 70)
````

---

## ➗ **1.4 Division - np.divide()**

**Purpose:** Divide corresponding elements

**Syntax:** `np.divide(array1, array2)` or `array1 / array2`

**Important:** Watch out for division by zero!

---


````python
# =============================================================================
# EXAMPLE 4: DIVISION
# =============================================================================
# Purpose: Demonstrate element-wise division

import numpy as np

# Create arrays of different shapes (broadcasting!)
a = np.array([[1, 2, 1],
              [2, 2, 3]])  # 2×3 array

b = np.array([3, 4, 5])   # 1D array (will broadcast)

print("=" * 70)
print("DIVISION: ELEMENT-WISE (with Broadcasting)")
print("=" * 70)
print()

print("📊 Array a (2×3):")
print(a)
print()

print("📊 Array b (1D):", b)
print()

# Perform division
result = np.divide(a, b)
# Or: result = a / b

print("➗ Result (a / b):")
print(result)
print()

print("💡 What happened (Broadcasting):")
print("   b is 'broadcast' to match a's shape:")
print("   [[1, 2, 1],  ÷  [[3, 4, 5],  =  [[0.33, 0.5, 0.2 ],")
print("    [2, 2, 3]]      [3, 4, 5]]      [0.67, 0.5, 0.6 ]]")
print()

# Integer vs Float division
a_int = np.array([10, 20, 30])
b_int = np.array([3, 4, 5])

print("🔢 Integer vs Float Division:")
print("   Regular division (/):", a_int / b_int, "(returns float)")
print("   Floor division  (//):", a_int // b_int, "(returns int)")
print()

# Real-world example
total = np.array([[100, 200], [300, 400]])
people = np.array([5, 10])
per_person = total / people

print("🌍 Real-world Example:")
print("   Total amount: $", total[0])
print("   People:       ", people)
print("   Per person:  $", per_person[0])
print()

print("=" * 70)
````

---

## 🔢 **1.5 Power - np.power()**

**Purpose:** Raise elements to powers

**Syntax:** `np.power(base, exponent)` or `base ** exponent`

---


````python
# =============================================================================
# EXAMPLE 5: POWER
# =============================================================================
# Purpose: Demonstrate element-wise power operation

import numpy as np

# Create arrays for base and exponent
a = [2, 2, 2, 2, 2]     # Base
b = [2, 3, 4, 5, 6]     # Exponent

print("=" * 70)
print("POWER: ELEMENT-WISE")
print("=" * 70)
print()

print("📊 Base (a):     ", a)
print("📊 Exponent (b): ", b)
print()

# Perform power operation
c = np.power(a, b)
# Or: c = np.array(a) ** np.array(b)

print("🔢 Result (a^b):", c)
print()

print("💡 What happened:")
print("   [2², 2³, 2⁴, 2⁵, 2⁶]")
print("   [4,  8,  16, 32, 64]")
print()

# More examples
base = np.array([1, 2, 3, 4, 5])

print("📐 Square each number:")
print("   ", base, "² =", base ** 2)
print()

print("📐 Cube each number:")
print("   ", base, "³ =", base ** 3)
print()

print("📐 Square root (power of 0.5):")
numbers = np.array([4, 9, 16, 25])
print("   ", numbers, "^0.5 =", numbers ** 0.5)
print()

# Real-world example
print("🌍 Real-world Example (Compound Interest):")
principal = np.array([1000, 2000, 3000])  # Initial investment
rate = 1.05  # 5% annual return
years = np.array([1, 2, 3])  # Years

# Formula: A = P * (1 + r)^t
final_amount = principal * (rate ** years)
print("   Principal:     $", principal)
print("   Rate:          5% per year")
print("   Years:         ", years)
print("   Final Amount: $", final_amount.round(2))
print()

print("=" * 70)
````

---

## 💡 **Key Concepts**

### **1. Element-Wise Operations**
All arithmetic operations are **element-wise** by default!

```python
[1, 2, 3] + [10, 20, 30] = [11, 22, 33]
```

### **2. Broadcasting**
NumPy automatically "stretches" smaller arrays to match larger ones:

```python
[[1, 2, 3],  +  [10, 20, 30]  =  [[11, 22, 33],
 [4, 5, 6]]                       [14, 25, 36]]
```

### **3. Scalar Operations**
Can operate on array with a single number:

```python
[1, 2, 3] * 10 = [10, 20, 30]
```

# 📊 **Part 2: Statistical Functions**

---

## 🤔 **Why Statistics Matter**

Statistics help us **understand data** by summarizing thousands of numbers into meaningful insights:

- 📊 **Central Tendency** - What's the "typical" value? (mean, median)
- 📏 **Spread** - How scattered is the data? (std, variance)
- 📈 **Distribution** - Where do most values fall? (percentiles)

**Every data scientist uses these daily!**

---

## 📋 **Key Statistical Functions**

| Function | What It Measures | Use When... |
|----------|------------------|-------------|
| `mean()` | Average value | General "typical" value |
| `median()` | Middle value | Data has outliers |
| `std()` | Spread around mean | Measure variability |
| `var()` | Spread squared | Mathematical calculations |
| `percentile()` | Value at position | Distribution analysis |

---

## 📊 **2.1 Core Statistical Measures**

Let's explore each function with real data!

---


````python
# =============================================================================
# STATISTICAL FUNCTIONS: MEAN, MEDIAN, STD, VARIANCE
# =============================================================================
# Purpose: Demonstrate core statistical calculations

import numpy as np

# Create a 2D array representing test scores
# - 3 students
# - 3 tests each
array = np.array([[4,  3,  2],    # Student 1
                  [10, 1,  0],    # Student 2
                  [5,  8,  24]])  # Student 3

print("=" * 70)
print("STATISTICAL ANALYSIS: UNDERSTANDING YOUR DATA")
print("=" * 70)
print()

print("📊 Test Scores Dataset:")
print(array)
print("   Shape:", array.shape, "→ 3 students × 3 tests")
print()

# ---------------------------------------------------------------------------
# MEDIAN: Middle value when sorted
# ---------------------------------------------------------------------------
median = np.median(array)
print("📍 1. MEDIAN (Middle Value):")
print("   Result:", median)
print()
print("   💡 What is median?")
print("      - Sort all values: [0, 1, 2, 3, 4, 5, 8, 10, 24]")
print("      - Take the middle: position 5 → value = 4")
print("      - Robust to outliers (24 doesn't skew it!)")
print()

# ---------------------------------------------------------------------------
# MEAN: Average value
# ---------------------------------------------------------------------------
mean = np.mean(array)
print("📊 2. MEAN (Average):")
print("   Result:", mean.round(2))
print()
print("   💡 What is mean?")
print("      - Sum all values: 4+3+2+10+1+0+5+8+24 = 57")
print("      - Divide by count: 57 / 9 = 6.33")
print("      - Affected by outliers (24 pulls it up!)")
print()

# ---------------------------------------------------------------------------
# STANDARD DEVIATION: Spread around the mean
# ---------------------------------------------------------------------------
std = np.std(array)
print("📏 3. STANDARD DEVIATION (Spread):")
print("   Result:", std.round(2))
print()
print("   💡 What is std?")
print("      - Measures how far values are from mean")
print("      - Low std = values close together")
print("      - High std = values spread out")
print("      - Our data: High std (", std.round(2), ") → very spread out!")
print()

# ---------------------------------------------------------------------------
# VARIANCE: Spread squared
# ---------------------------------------------------------------------------
variance = np.var(array)
print("📐 4. VARIANCE (Spread²):")
print("   Result:", variance.round(2))
print()
print("   💡 What is variance?")
print("      - Square of standard deviation")
print("      - variance = std²")
print("      - Check:", std.round(2), "² =", (std**2).round(2))
print("      - Used in many statistical formulas")
print()

print("=" * 70)
````

---

## 🎯 **Understanding Mean vs Median**

### **When to use Mean:**
✅ Data is normally distributed (bell curve)
✅ No extreme outliers
✅ You want mathematical properties (e.g., sum/count)

**Example:** Average salary in a small team

### **When to use Median:**
✅ Data has outliers
✅ Data is skewed
✅ You want a "typical" value

**Example:** Median home price (billionaire's mansion shouldn't count!)

### **Visual Example:**

**Salaries: [$30K, $35K, $40K, $45K, $50K, $1M]**

- **Mean:** $200K (misleading! Pulled up by $1M)
- **Median:** $42.5K (more representative of "typical" salary)

---

## 📏 **Understanding Standard Deviation**

**Standard Deviation (std)** tells you how spread out your data is.

### **Low std - Data clustered together:**
```
Heights: [170cm, 171cm, 169cm, 170cm, 171cm]
std ≈ 0.8cm → Everyone similar height!
```

### **High std - Data spread apart:**
```
Heights: [150cm, 190cm, 160cm, 185cm, 165cm]
std ≈ 15cm → Very different heights!
```

### **Real-world use:**
- **Stock volatility:** High std = risky stock
- **Quality control:** Low std = consistent product
- **Test scores:** Low std = fair test, High std = mixed abilities

---

## 📈 **2.2 Percentiles - Understanding Distribution**

**Percentile:** Value below which X% of data falls

**Examples:**
- **50th percentile** = Median (50% below, 50% above)
- **25th percentile** = 1st Quartile (25% below)
- **75th percentile** = 3rd Quartile (75% below)
- **90th percentile** = 90% of values are below this

---


````python
# =============================================================================
# PERCENTILES: UNDERSTANDING DATA DISTRIBUTION
# =============================================================================
# Purpose: Demonstrate percentile calculations

import numpy as np

# Use the same array from before
array = np.array([[4, 3, 2], [10, 1, 0], [5, 8, 24]])

print("=" * 70)
print("PERCENTILES: WHERE DOES YOUR DATA FALL?")
print("=" * 70)
print()

print("📊 Dataset:", array.flatten())
print("   Sorted:  ", np.sort(array.flatten()))
print()

# Calculate various percentiles
p25 = np.percentile(array, 25)
p50 = np.percentile(array, 50)
p75 = np.percentile(array, 75)
p90 = np.percentile(array, 90)

print("📈 Percentile Analysis:")
print()
print("   25th percentile:", p25, "→ 25% of values ≤", p25)
print("   50th percentile:", p50, "→ 50% of values ≤", p50, "(MEDIAN!)")
print("   75th percentile:", p75, "→ 75% of values ≤", p75)
print("   90th percentile:", p90, "→ 90% of values ≤", p90)
print()

print("💡 What this means:")
print("   - Half the scores are below", p50)
print("   - Top 25% scored above", p75)
print("   - Top 10% scored above", p90)
print()

# Real-world example
print("🌍 Real-world Example (Exam Scores):")
exam_scores = np.array([45, 67, 23, 89, 91, 56, 78, 88, 92, 45, 67, 99, 100])
print("   Scores:", exam_scores)
print()
print("   25th percentile:", np.percentile(exam_scores, 25), "→ Bottom 25%")
print("   50th percentile:", np.percentile(exam_scores, 50), "→ Median")
print("   75th percentile:", np.percentile(exam_scores, 75), "→ Top 25%")
print()
print("   🎓 If you scored 78, you're at the",
      ((exam_scores <= 78).sum() / len(exam_scores) * 100).round(0),
      "percentile!")
print()

print("=" * 70)
````

## 🎯 **Statistical Functions with Axis**

You can calculate statistics **along specific dimensions**!

**axis=None** (default) - Across entire array
**axis=0** - Down columns (across rows)
**axis=1** - Across rows (across columns)

---


````python
# =============================================================================
# STATISTICS WITH AXIS
# =============================================================================
# Purpose: Show how to calculate stats along specific dimensions

import numpy as np

# Student scores: 3 students × 3 tests
scores = np.array([[85, 90, 78],   # Student 1
                   [92, 88, 95],   # Student 2
                   [70, 75, 80]])  # Student 3

print("=" * 70)
print("STATISTICS WITH AXIS: DIFFERENT PERSPECTIVES")
print("=" * 70)
print()

print("📊 Student Scores (3 students × 3 tests):")
print(scores)
print()

# Overall statistics
print("📈 Overall Statistics:")
print("   Overall mean:  ", scores.mean().round(2))
print("   Overall median:", np.median(scores))
print()

# Statistics per student (axis=1: across columns)
print("👤 Per Student (axis=1):")
print("   Student 1 average:", scores[0].mean().round(2))
print("   Student 2 average:", scores[1].mean().round(2))
print("   Student 3 average:", scores[2].mean().round(2))
print("   Or all at once:   ", scores.mean(axis=1).round(2))
print()

# Statistics per test (axis=0: down rows)
print("📝 Per Test (axis=0):")
print("   Test 1 average:", scores[:, 0].mean().round(2))
print("   Test 2 average:", scores[:, 1].mean().round(2))
print("   Test 3 average:", scores[:, 2].mean().round(2))
print("   Or all at once:", scores.mean(axis=0).round(2))
print()

print("=" * 70)
````

# 🔤 **Part 3: String Functions**

---

## 🤔 **Why String Functions in NumPy?**

While NumPy is primarily for numbers, it also provides **efficient string operations**!

**Use cases:**
- 📝 Text data preprocessing
- 🏷️ Label manipulation
- 📊 Cleaning categorical data
- 🔄 Batch text transformations

**All operations are vectorized = FAST!** ⚡

---


````python
# =============================================================================
# STRING CONCATENATION
# =============================================================================
# Purpose: Demonstrate element-wise string concatenation

import numpy as np

x = np.array(['Hello', 'World'])
y = np.array(['Welcome', 'Learners'])

print("=" * 70)
print("STRING CONCATENATION")
print("=" * 70)
print()

print("Array x:", x)
print("Array y:", y)
print()

result = np.char.add(x, y)
print("Concatenated:", result)
print()

print("=" * 70)
````

````python
# =============================================================================
# STRING REPLACEMENT
# =============================================================================
# Purpose: Demonstrate substring replacement

import numpy as np

str_original = "Hello How Are You"

print("=" * 70)
print("STRING REPLACEMENT")
print("=" * 70)
print()

print("Original:", str_original)

result = np.char.replace(str_original, 'Hello', 'Hi')
print("Modified:", result)
print()

print("=" * 70)
````

````python
# =============================================================================
# CASE CONVERSION
# =============================================================================
# Purpose: Demonstrate case conversion

import numpy as np

print("=" * 70)
print("CASE CONVERSION")
print("=" * 70)
print()

# Lowercase to UPPERCASE
str_lower = "hello how are you"
print("Original (lowercase):", str_lower)
result_upper = np.char.upper(str_lower)
print("UPPERCASE:          ", result_upper)
print()

# UPPERCASE to lowercase
str_upper = "GREETINGS OF THE DAY"
print("Original (UPPERCASE):", str_upper)
result_lower = np.char.lower(str_upper)
print("lowercase:          ", result_lower)
print()

print("=" * 70)
````
