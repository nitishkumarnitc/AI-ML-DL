## ⚡ **Key Features of Pandas**



## 🆚 **Pandas vs NumPy**

| Aspect | NumPy | Pandas |
|--------|-------|--------|
| **Data Type** | Homogeneous (all same type) | Heterogeneous (mixed types) |
| **Structure** | Arrays (no labels) | Series & DataFrame (labeled) |
| **Best For** | Numerical computations | Data analysis & manipulation |
| **Indexing** | Integer positions only | Labels + integers |
| **Missing Data** | Manual handling | Built-in NaN support |
| **Performance** | Faster for math | Easier for real-world data |

**Bottom line:** NumPy for heavy number crunching, Pandas for everything else!

## 📦 **Pandas Data Structures**

Pandas has **TWO** main data structures:

### **1. Series - 1D Labeled Array** 📍

Think: A single column from a spreadsheet

```python
s = pd.Series([10, 20, 30, 40])
```

**Properties:**
- One-dimensional
- Homogeneous data (one type)
- Labeled index
- Like a dictionary that remembers order

### **2. DataFrame - 2D Labeled Table** 📊

Think: An entire spreadsheet

```python
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['NYC', 'LA', 'Chicago']
})
```

**Properties:**
- Two-dimensional (rows & columns)
- Heterogeneous columns (different types)
- Labeled rows and columns
- Collection of Series

# 📍 **Part 2: Pandas Series - Labeled 1D Data**

---

## 🤔 **What is a Series?**

**A Series is a 1D labeled array** that can hold any data type.

**Components:**
- **Values** - The actual data (like a NumPy array)
- **Index** - Labels for each value (like dictionary keys)

**Analogy:** A dictionary that remembers order + NumPy array capabilities

```
Index    Values
  A   →    10
  B   →    20
  C   →    30
  D   →    40
```

---

## 🔨 **Creating a Series: Multiple Methods**

You can create Series from:
- ✅ Lists
- ✅ NumPy arrays
- ✅ Dictionaries
- ✅ Scalar values
- ✅ Other Series

---


````
# =============================================================================
# CREATING PANDAS SERIES - ALL METHODS
# =============================================================================
# Purpose: Demonstrate multiple ways to create a Series

import pandas as pd
import numpy as np

print("=" * 70)
print("CREATING PANDAS SERIES")
print("=" * 70)
print()

# ---------------------------------------------------------------------------
# Method 1: From a List (most common)
# ---------------------------------------------------------------------------
print("📋 Method 1: From a List")
data = [1, 2, 3, 4, 5]
series = pd.Series(data)

print("   Data:", data)
print("   Series:")
print(series)
print()
print("   💡 Index automatically created (0, 1, 2, 3, 4)")
print()

# ---------------------------------------------------------------------------
# Method 2: From a List with Custom Index
# ---------------------------------------------------------------------------
print("🏷️ Method 2: List with Custom Index")
index = ['a', 'b', 'c', 'd', 'e']
series_with_index = pd.Series(data, index=index)

print("   Series with custom labels:")
print(series_with_index)
print()
print("   💡 Now we can use labels: 'a', 'b', 'c', etc.")
print()

# ---------------------------------------------------------------------------
# Method 3: From a Dictionary (index from keys)
# ---------------------------------------------------------------------------
print("📖 Method 3: From a Dictionary")
data_dict = {'a': 1, 'b': 2, 'c': 3, 'd': 4, 'e': 5}
series_from_dict = pd.Series(data_dict)

print("   Dictionary:", data_dict)
print("   Series:")
print(series_from_dict)
print()
print("   💡 Keys become index, values become data")
print()

# ---------------------------------------------------------------------------
# Method 4: From NumPy Array
# ---------------------------------------------------------------------------
print("🔢 Method 4: From NumPy Array")
numpy_array = np.array([10, 20, 30, 40, 50])
series_from_numpy = pd.Series(numpy_array, index=['Mon', 'Tue', 'Wed', 'Thu', 'Fri'])

print("   NumPy array:", numpy_array)
print("   Series (days of week):")
print(series_from_numpy)
print()

# ---------------------------------------------------------------------------
# Method 5: From a Scalar Value
# ---------------------------------------------------------------------------
print("📍 Method 5: From a Scalar Value")
series_scalar = pd.Series(100, index=['A', 'B', 'C'])

print("   Series (repeated value):")
print(series_scalar)
print()
print("   💡 Scalar value repeated for all indices")
print()

# ---------------------------------------------------------------------------
# Real-world example
# ---------------------------------------------------------------------------
print("🌍 Real-World Example: Student Grades")
grades = pd.Series({
    'Alice': 92,
    'Bob': 85,
    'Charlie': 78,
    'Diana': 95,
    'Eve': 88
})

print(grades)
print()

print("=" * 70)
````

## 🎯 **Accessing Series Data**

**Two ways to access elements:**
1. **Label-based** - Using index labels
2. **Position-based** - Using integer positions

---


````
# =============================================================================
# ACCESSING SERIES DATA
# =============================================================================
# Purpose: Demonstrate different ways to access Series elements

import pandas as pd

# Create sample Series
series_with_index = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e'])

print("=" * 70)
print("ACCESSING SERIES DATA")
print("=" * 70)
print()

print("📊 Our Series:")
print(series_with_index)
print()

# ---------------------------------------------------------------------------
# Label-based access
# ---------------------------------------------------------------------------
print("🏷️ Label-Based Access:")
print()

# Single element by label
print("   series['b'] =", series_with_index['b'])
print("   → Get value at label 'b'")
print()

# Multiple elements by labels
print("   series[['a', 'c', 'e']] =")
print(series_with_index[['a', 'c', 'e']])
print()

# ---------------------------------------------------------------------------
# Position-based access
# ---------------------------------------------------------------------------
print("📍 Position-Based Access:")
print()

# Single element by position
print("   series[2] =", series_with_index[2])
print("   → Get value at position 2 (third element)")
print()

# Slice by position
print("   series[1:4] =")
print(series_with_index[1:4])
print("   → Elements at positions 1, 2, 3")
print()

# ---------------------------------------------------------------------------
# Using .loc and .iloc
# ---------------------------------------------------------------------------
print("🎯 Using .loc (label) and .iloc (position):")
print()

print("   .loc['b'] =", series_with_index.loc['b'], "(label-based)")
print("   .iloc[1] =", series_with_index.iloc[1], "(position-based)")
print()

print("   💡 .loc uses LABELS, .iloc uses POSITIONS")
print()

# ---------------------------------------------------------------------------
# Real-world example
# ---------------------------------------------------------------------------
print("🌍 Real-World Example:")
stock_prices = pd.Series({
    'AAPL': 175.50,
    'GOOGL': 140.25,
    'MSFT': 380.75,
    'AMZN': 145.30
})

print("   Stock Prices:")
print(stock_prices)
print()

print("   Apple stock:", stock_prices['AAPL'])
print("   First stock:", stock_prices.iloc[0])
print("   Tech giants:", stock_prices[['AAPL', 'GOOGL', 'MSFT']])
print()

print("=" * 70)
````

## 📊 **Series: Basic Information & Statistics**

Every Series has powerful built-in methods to explore and understand your data!

---


````
# =============================================================================
# SERIES: BASIC INFORMATION & STATISTICS
# =============================================================================
# Purpose: Demonstrate exploration methods for Series

import pandas as pd
import numpy as np

# Create sample Series
series = pd.Series([15, 22, 18, 30, 25, 18, 40, 35, 22, 28],
                    index=['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J'])

print("=" * 70)
print("SERIES: BASIC INFORMATION")
print("=" * 70)
print()

print("📊 Our Series:")
print(series)
print()

# ---------------------------------------------------------------------------
# Head and Tail
# ---------------------------------------------------------------------------
print("🔝 First Few Elements:")
print(series.head(3))
print()

print("🔚 Last Few Elements:")
print(series.tail(3))
print()

# ---------------------------------------------------------------------------
# Shape and Size
# ---------------------------------------------------------------------------
print("📏 Shape & Size:")
print(f"   Shape: {series.shape} (number of elements)")
print(f"   Size:  {series.size} elements")
print(f"   dtype: {series.dtype}")
print()

# ---------------------------------------------------------------------------
# Descriptive Statistics
# ---------------------------------------------------------------------------
print("📈 Descriptive Statistics:")
print(series.describe())
print()

print("💡 Quick Stats:")
print(f"   Mean:   {series.mean():.2f}")
print(f"   Median: {series.median():.2f}")
print(f"   Std Dev: {series.std():.2f}")
print(f"   Min:    {series.min()}")
print(f"   Max:    {series.max()}")
print(f"   Sum:    {series.sum()}")
print()

# ---------------------------------------------------------------------------
# Unique Values
# ---------------------------------------------------------------------------
print("🔍 Unique Values:")
print(f"   Unique values: {series.unique()}")
print(f"   Number of unique: {series.nunique()}")
print(f"   Value counts:")
print(series.value_counts())
print()

# ---------------------------------------------------------------------------
# Real-world example
# ---------------------------------------------------------------------------
print("🌍 Real-World Example: Test Scores")
scores = pd.Series([85, 92, 78, 92, 88, 85, 95, 78, 90, 85],
                    index=[f'Student_{i}' for i in range(1, 11)])

print(scores)
print()
print(f"Class Average: {scores.mean():.1f}")
print(f"Highest Score: {scores.max()}")
print(f"Most Common Score: {scores.mode()[0]}")
print()

print("=" * 70)
````

# 🔄 **Part 3: Series Operations & Transformations**

---

## 🎯 **Powerful Series Operations**

Series support:
- ✅ Arithmetic operations (element-wise)
- ✅ Apply custom functions
- ✅ Mapping and replacement
- ✅ Sorting
- ✅ Missing data handling

---


````
# =============================================================================
# SERIES OPERATIONS & TRANSFORMATIONS
# =============================================================================
# Purpose: Demonstrate various operations on Series

import pandas as pd
import numpy as np

# Create sample Series
s1 = pd.Series([10, 20, 30, 40, 50], index=['A', 'B', 'C', 'D', 'E'])
s2 = pd.Series([1, 2, 3, 4, 5], index=['A', 'B', 'C', 'D', 'E'])

print("=" * 70)
print("SERIES OPERATIONS & TRANSFORMATIONS")
print("=" * 70)
print()

# ---------------------------------------------------------------------------
# Arithmetic Operations
# ---------------------------------------------------------------------------
print("➕ Arithmetic Operations:")
print()

print("Series 1:")
print(s1)
print()

print("Series 2:")
print(s2)
print()

print("Addition (s1 + s2):")
print(s1 + s2)
print()

print("Multiplication (s1 * s2):")
print(s1 * s2)
print()

print("Scalar operations (s1 * 2):")
print(s1 * 2)
print()

# ---------------------------------------------------------------------------
# Apply Functions
# ---------------------------------------------------------------------------
print("🔧 Apply Custom Functions:")
print()

# Square each element
squared = s1.apply(lambda x: x**2)
print("Squared values:")
print(squared)
print()

# Custom function
def categorize(x):
    if x < 30:
        return 'Low'
    elif x < 50:
        return 'Medium'
    else:
        return 'High'

categories = s1.apply(categorize)
print("Categorized:")
print(categories)
print()

# ---------------------------------------------------------------------------
# Mapping Values
# ---------------------------------------------------------------------------
print("🗺️ Map Values:")
print()

grades = pd.Series([1, 2, 3, 4, 5])
grade_map = {1: 'F', 2: 'D', 3: 'C', 4: 'B', 5: 'A'}
letters = grades.map(grade_map)

print("Numeric grades:", grades.values)
print("Letter grades:")
print(letters)
print()

# ---------------------------------------------------------------------------
# Sorting
# ---------------------------------------------------------------------------
print("📊 Sorting:")
print()

unsorted = pd.Series([30, 10, 50, 20, 40], index=['C', 'A', 'E', 'B', 'D'])
print("Original:")
print(unsorted)
print()

print("Sort by values:")
print(unsorted.sort_values())
print()

print("Sort by index:")
print(unsorted.sort_index())
print()

# ---------------------------------------------------------------------------
# Missing Data
# ---------------------------------------------------------------------------
print("🧹 Handling Missing Data:")
print()

series_with_nan = pd.Series([10, np.nan, 30, np.nan, 50])
print("Series with NaN:")
print(series_with_nan)
print()

print("Check for missing:")
print(series_with_nan.isnull())
print()

print("Fill missing with 0:")
print(series_with_nan.fillna(0))
print()

print("Drop missing:")
print(series_with_nan.dropna())
print()

# ---------------------------------------------------------------------------
# Real-world example
# ---------------------------------------------------------------------------
print("🌍 Real-World: Temperature Conversion")
celsius = pd.Series([0, 10, 20, 30, 40],
                     index=['Mon', 'Tue', 'Wed', 'Thu', 'Fri'])
fahrenheit = celsius.apply(lambda c: (c * 9/5) + 32)

print("Celsius:")
print(celsius)
print()
print("Fahrenheit:")
print(fahrenheit)
print()

print("=" * 70)
````

### __3.4 Querying a Series__
Selecting and filtering data based on specific conditions is an essential aspect of querying a Pandas Series.

The following examples illustrate common querying operations that can be applied to a Pandas Series:

````
import pandas as pd

# Create a Pandas Series
data = {'a': 10, 'b': 20, 'c': 30, 'd': 40, 'e': 50}
series = pd.Series(data)

# Select elements greater than 30
selected_greater_than_30 = series[series > 30]

# Select elements equal to 20
selected_equal_to_20 = series[series == 20]

# Select elements not equal to 40
selected_not_equal_to_40 = series[series != 40]

# Select elements based on multiple conditions
selected_multiple_conditions = series[(series > 20) & (series < 50)]

# Select elements based on a list of values
selected_by_list = series[series.isin([20, 40, 60])]

# Select elements using string methods (if applicable)
string_series = pd.Series(['apple', 'banana', 'cherry', 'date', 'elderberry'])
selected_by_string_method = string_series[string_series.str.startswith('b')]

# Query based on index labels
selected_by_index_labels = series.loc[['a', 'c', 'e']]

# Query based on numeric position
selected_by_numeric_position = series.iloc[1:4]

print("=" * 70)
print("DEMONSTRATION: ALL QUERYING METHODS")
print("=" * 70)
print()

print("Original Series:")
print(series)
print("\n✅ Selected > 30:")
print(selected_greater_than_30)
print("\n✅ Selected == 20:")
print(selected_equal_to_20)
print("\n✅ Selected != 40:")
print(selected_not_equal_to_40)
print("\n✅ Multiple conditions (20 < x < 50):")
print(selected_multiple_conditions)
print("\n✅ In list [20, 40, 60]:")
print(selected_by_list)
print("\n✅ String method (starts with 'b'):")
print(selected_by_string_method)
print("\n✅ By labels ['a', 'c', 'e']:")
print(selected_by_index_labels)
print("\n✅ By positions [1:4]:")
print(selected_by_numeric_position)
print()

print("=" * 70)
````
