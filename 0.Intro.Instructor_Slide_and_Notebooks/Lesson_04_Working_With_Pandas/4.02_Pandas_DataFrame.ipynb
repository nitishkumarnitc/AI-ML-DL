# 📊 **Part 1: Understanding DataFrames**

---

## 🤔 **What is a DataFrame?**

**A DataFrame is a 2D labeled data structure** with columns that can hold different types.

**Structure:**
```
        Column1  Column2  Column3
Row1       A       100     True
Row2       B       200     False
Row3       C       300     True
```

**Think of it as:**
- 📋 A table with named rows and columns
- 📊 An Excel spreadsheet in Python
- 🗃️ A collection of Series sharing the same index

---

## 🎯 **Key Features of DataFrames**

### **1. Tabular Structure** 📊
- Organized in rows and columns
- Like a spreadsheet or SQL table
- Easy to visualize and understand

### **2. Labeled Axes** 🏷️
- **Rows:** Index labels (0, 1, 2... or custom)
- **Columns:** Column names ('Name', 'Age', etc.)
- Access by name OR position

### **3. Heterogeneous Data** 🎨
- **Different types per column!**
- Column 1: strings, Column 2: integers, Column 3: floats
- Each column is homogeneous (same type within)

### **4. Versatility** 🔄
- Read from: CSV, Excel, SQL, JSON, HTML, clipboard
- Write to: CSV, Excel, SQL, JSON, HTML, LaTeX
- In-memory manipulation

### **5. Data Alignment** ⚙️
- Automatic alignment in operations
- Handles missing data (NaN) gracefully
- Smart joining and merging

---

## 📐 **DataFrame Structure Explained**

```python
        Name    Age   Salary    City
0      Alice    25    50000     NYC
1      Bob      30    60000     LA
2      Charlie  22    45000     Chicago
3      Diana    28    55000     Boston

↑       ↑       ↑      ↑         ↑
Index  Col1    Col2   Col3      Col4
```

**Components:**
- **Index:** Row labels (0, 1, 2, 3)
- **Columns:** Column names (Name, Age, Salary, City)
- **Values:** The actual data
- **Shape:** (4 rows, 4 columns) = (4, 4)

---

## 🆚 **Series vs DataFrame**

| Aspect | Series | DataFrame |
|--------|--------|-----------|
| **Dimensions** | 1D | 2D |
| **Structure** | Single column | Multiple columns |
| **Analogy** | A list with labels | A table/spreadsheet |
| **Data types** | One type | Different per column |
| **Think of it as** | One column | Collection of Series |

**Relationship:** A DataFrame is multiple Series stacked together!

## 🔨 **Creating DataFrames: Multiple Methods**

You can create DataFrames from almost any data source!

---


````python
# =============================================================================
# METHOD 1: FROM DICTIONARY (MOST COMMON!)
# =============================================================================
# Purpose: Create DataFrame from dictionary - each key becomes a column

import pandas as pd

print("=" * 70)
print("METHOD 1: FROM DICTIONARY")
print("=" * 70)
print()

# Dictionary: keys = column names, values = column data
data_dict = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana'],
    'Age': [25, 30, 22, 28],
    'Salary': [50000, 60000, 45000, 55000],
    'City': ['NYC', 'LA', 'Chicago', 'Boston']
}

df_dict = pd.DataFrame(data_dict)

print("📊 DataFrame from Dictionary:")
print(df_dict)
print()
print(f"Shape: {df_dict.shape} (4 rows × 4 columns)")
print()

# With custom index
df_custom_index = pd.DataFrame(data_dict, index=['Emp1', 'Emp2', 'Emp3', 'Emp4'])
print("📋 With Custom Index:")
print(df_custom_index)
print()

print("=" * 70)
print()
````

````python
# =============================================================================
# METHOD 2: FROM LIST OF LISTS
# =============================================================================
# Purpose: Create DataFrame from nested lists - each inner list is a row

import pandas as pd

print("=" * 70)
print("METHOD 2: FROM LIST OF LISTS")
print("=" * 70)
print()

# List of lists - each inner list is a row
data_list = [
    ['Alice', 25, 50000, 'NYC'],
    ['Bob', 30, 60000, 'LA'],
    ['Charlie', 22, 45000, 'Chicago'],
    ['Diana', 28, 55000, 'Boston']
]

# Must specify column names
columns = ['Name', 'Age', 'Salary', 'City']

df_list = pd.DataFrame(data_list, columns=columns)

print("📊 DataFrame from Lists:")
print(df_list)
print()

print("=" * 70)
print()
````

````python
# =============================================================================
# METHOD 3: FROM NUMPY ARRAY
# =============================================================================
# Purpose: Convert NumPy array to DataFrame

import pandas as pd
import numpy as np

print("=" * 70)
print("METHOD 3: FROM NUMPY ARRAY")
print("=" * 70)
print()

# NumPy array (2D)
data_array = np.array([
    ['Alice', 25, 50000],
    ['Bob', 30, 60000],
    ['Charlie', 22, 45000]
])

columns = ['Name', 'Age', 'Salary']
df_array = pd.DataFrame(data_array, columns=columns)

print("📊 DataFrame from NumPy Array:")
print(df_array)
print()
print("💡 Note: All data becomes strings by default!")
print("   Use dtype parameter to specify types")
print()

# Random data example
random_data = np.random.randint(0, 100, size=(5, 3))
df_random = pd.DataFrame(random_data, columns=['A', 'B', 'C'])

print("🎲 DataFrame from Random Array:")
print(df_random)
print()

print("=" * 70)
print()
````

````python
# =============================================================================
# METHOD 4: FROM CSV FILE (REAL-WORLD!)
# =============================================================================
# Purpose: Read data from CSV file - most common in practice!

import pandas as pd

print("=" * 70)
print("METHOD 4: FROM CSV FILE")
print("=" * 70)
print()

# Sample: Create a CSV-like DataFrame for demonstration
sample_data = {
    'Product': ['Laptop', 'Phone', 'Tablet', 'Monitor', 'Keyboard'],
    'Price': [1200, 800, 500, 300, 50],
    'Stock': [15, 30, 25, 40, 100],
    'Category': ['Electronics', 'Electronics', 'Electronics', 'Accessories', 'Accessories']
}

df_products = pd.DataFrame(sample_data)

print("📊 Sample Product Data (like from CSV):")
print(df_products)
print()

print("💡 In practice, load CSV files:")
print("   df = pd.read_csv('data.csv')")
print("   df = pd.read_csv('data.csv', index_col=0)  # First column as index")
print()

print("=" * 70)
print()
````

````python
# =============================================================================
# METHOD 5: FROM LIST OF DICTIONARIES
# =============================================================================
# Purpose: Each dictionary becomes a row

import pandas as pd

print("=" * 70)
print("METHOD 5: FROM LIST OF DICTIONARIES")
print("=" * 70)
print()

# List where each element is a dictionary (row)
data_dicts = [
    {'Name': 'Alice', 'Age': 25, 'City': 'NYC'},
    {'Name': 'Bob', 'Age': 30, 'City': 'LA'},
    {'Name': 'Charlie', 'Age': 22, 'City': 'Chicago'}
]

df_list_dicts = pd.DataFrame(data_dicts)

print("📊 DataFrame from List of Dicts:")
print(df_list_dicts)
print()

# Handles missing values automatically
data_incomplete = [
    {'Name': 'Alice', 'Age': 25, 'City': 'NYC'},
    {'Name': 'Bob', 'City': 'LA'},  # Missing Age!
    {'Name': 'Charlie', 'Age': 22}  # Missing City!
]

df_incomplete = pd.DataFrame(data_incomplete)

print("🔧 Handles Missing Values:")
print(df_incomplete)
print("   (NaN = Not a Number/Missing)")
print()

print("=" * 70)
````

## 📋 **Quick Reference: Creation Methods**

| Method | Use When... | Example |
|--------|-------------|---------|
| **Dictionary** | Creating from scratch | `pd.DataFrame({'A': [1,2], 'B': [3,4]})` |
| **List of Lists** | Row-oriented data | `pd.DataFrame([[1,2], [3,4]], columns=['A','B'])` |
| **NumPy Array** | Converting numeric data | `pd.DataFrame(array, columns=cols)` |
| **CSV File** | Real-world data | `pd.read_csv('file.csv')` |
| **Excel** | Excel files | `pd.read_excel('file.xlsx')` |
| **SQL** | Database queries | `pd.read_sql(query, connection)` |
| **JSON** | Web APIs | `pd.read_json('data.json')` |

**Most common in practice:** `pd.read_csv()` and dictionary creation

# 🎯 **Part 2: Accessing DataFrame Data**

Multiple ways to get exactly what you need!

---


````python
# =============================================================================
# ACCESSING DATAFRAME DATA - COMPLETE GUIDE
# =============================================================================
# Purpose: Demonstrate all methods to access DataFrame data

import pandas as pd

# Create sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age': [25, 30, 22, 28, 35],
    'Salary': [50000, 60000, 45000, 55000, 70000],
    'City': ['NYC', 'LA', 'Chicago', 'Boston', 'Seattle']
}

df = pd.DataFrame(data)

print("=" * 70)
print("ACCESSING DATAFRAME DATA")
print("=" * 70)
print()

print("📊 Our DataFrame:")
print(df)
print()

# Column access
print("📋 1. ACCESSING COLUMNS:")
print()
print("Single column (returns Series):")
print(df['Name'])
print()

print("Multiple columns (returns DataFrame):")
print(df[['Name', 'Salary']])
print()

# Row access
print("📍 2. ACCESSING ROWS:")
print()
print("By position (.iloc):")
print(df.iloc[0])  # First row
print()

print("By label (.loc):")
print(df.loc[2])  # Row with index 2
print()

print("Slice rows:")
print(df.iloc[1:4])  # Rows 1, 2, 3
print()

# Cell access
print("🎯 3. ACCESSING CELLS:")
print()
print(f"Cell at row 0, column 'Name': {df.at[0, 'Name']}")
print(f"Cell at position (1, 2): {df.iat[1, 2]}")
print()

# Boolean indexing
print("🔍 4. BOOLEAN INDEXING (FILTERING):")
print()
print("People with Salary > 50000:")
print(df[df['Salary'] > 50000])
print()

print("People in NYC or LA:")
print(df[df['City'].isin(['NYC', 'LA'])])
print()

print("Multiple conditions (Age > 25 AND Salary > 50000):")
print(df[(df['Age'] > 25) & (df['Salary'] > 50000)])
print()

print("=" * 70)
````

## 📊 **DataFrame Basics: Essential Methods**

Every DataFrame has powerful methods to explore and understand your data!

---


````python
# =============================================================================
# DATAFRAME BASICS - EXPLORATION METHODS
# =============================================================================
# Purpose: Demonstrate essential DataFrame exploration methods

import pandas as pd

# Create sample DataFrame
data = {
    'Product': ['Laptop', 'Phone', 'Tablet', 'Monitor', 'Keyboard', 'Mouse', 'Speaker', 'Headphones'],
    'Price': [1200, 800, 500, 300, 50, 25, 150, 100],
    'Stock': [15, 30, 25, 40, 100, 150, 60, 80],
    'Category': ['Electronics', 'Electronics', 'Electronics', 'Accessories', 'Accessories', 'Accessories', 'Audio', 'Audio']
}

df = pd.DataFrame(data)

print("=" * 70)
print("DATAFRAME BASICS")
print("=" * 70)
print()

# Head and tail
print("🔝 First 3 rows:")
print(df.head(3))
print()

print("🔚 Last 2 rows:")
print(df.tail(2))
print()

# Info
print("ℹ️ DataFrame Info:")
df.info()
print()

# Shape and size
print(f"📏 Shape: {df.shape} (rows, columns)")
print(f"📦 Size: {df.size} total elements")
print(f"📋 Columns: {list(df.columns)}")
print(f"🔢 Index: {list(df.index)}")
print()

# Describe
print("📈 Descriptive Statistics (numeric columns only):")
print(df.describe())
print()

print("=" * 70)
````

# 📊 **Part 3: Statistical Operations**

DataFrames make statistics easy and powerful!

---


````python
# =============================================================================
# STATISTICAL OPERATIONS IN PANDAS
# =============================================================================
# Purpose: Demonstrate statistical analysis with DataFrames

import pandas as pd
import numpy as np

# Create sample DataFrame
np.random.seed(42)
data = {
    'Student': [f'Student_{i}' for i in range(1, 11)],
    'Math': np.random.randint(60, 100, 10),
    'English': np.random.randint(65, 100, 10),
    'Science': np.random.randint(55, 95, 10),
    'History': np.random.randint(60, 95, 10)
}

df_scores = pd.DataFrame(data)

print("=" * 70)
print("STATISTICAL OPERATIONS")
print("=" * 70)
print()

print("📊 Student Scores:")
print(df_scores)
print()

# Descriptive statistics
print("📈 DESCRIPTIVE STATISTICS:")
print()

print("Mean scores per subject:")
print(df_scores[['Math', 'English', 'Science', 'History']].mean())
print()

print("Median scores:")
print(df_scores[['Math', 'English', 'Science', 'History']].median())
print()

print("Standard deviation:")
print(df_scores[['Math', 'English', 'Science', 'History']].std())
print()

print("Min and Max:")
print("Min:", df_scores[['Math', 'English', 'Science', 'History']].min().values)
print("Max:", df_scores[['Math', 'English', 'Science', 'History']].max().values)
print()

# Correlation
print("🔗 CORRELATION ANALYSIS:")
print()

correlation = df_scores[['Math', 'English', 'Science', 'History']].corr()
print("Correlation matrix:")
print(correlation)
print()

print("💡 Interpretation:")
print("   1.0 = perfect positive correlation")
print("   0.0 = no correlation")
print("   -1.0 = perfect negative correlation")
print()

# Aggregations
print("🎯 AGGREGATIONS:")
print()

# Sum per column
print("Total points per subject:")
print(df_scores[['Math', 'English', 'Science', 'History']].sum())
print()

# Count
print(f"Number of students: {df_scores['Student'].count()}")
print()

# Custom aggregations
print("Summary per subject:")
summary = df_scores[['Math', 'English', 'Science', 'History']].agg(['mean', 'median', 'std', 'min', 'max'])
print(summary)
print()

print("=" * 70)
````

### __2.1 Mean, Median, and Standard Deviation__

````python
import pandas as pd

# Create a sample DataFrame with numeric columns
data = {'Numeric_column1': [5, 15, 8],
        'Numeric_column2': [10, 20, 30],
        'Numeric_column3': [100, 200, 300]}

df = pd.DataFrame(data)

# Calculate mean, median, and standard deviation
mean_value = df.mean()
median_value = df.median()
std_deviation = df.std()

print("Mean:\n", mean_value)
print("\nMedian:\n", median_value)
print("\nStandard deviation:\n", std_deviation)
````

### __2.2 Correlation Analysis__
The corr() function generates a correlation matrix, indicating how variables relate to each other.

Values closer to 1 or -1 imply a stronger correlation, while values near 0 suggest a weaker correlation.

````python
import pandas as pd

# Create a sample DataFrame with numeric columns
data = {'Numeric_column1': [5, 15, 8],
        'Numeric_column2': [10, 20, 30],
        'Numeric_column3': [100, 200, 300]}

df = pd.DataFrame(data)

# Compute correlation matrix
correlation_matrix = df.corr()

print("Correlation matrix:\n", correlation_matrix)
````

#### __Value Counts__
The value_counts() function tallies the occurrences of unique values in a categorical column, aiding in understanding the distribution of categorical data.

````python
import pandas as pd

# Create a sample DataFrame with a category column
data = {'Category': ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'A', 'B', 'C']}
df = pd.DataFrame(data)

# Count occurrences of unique values in the category column
value_counts = df['Category'].value_counts()

print("Value counts:\n", value_counts)
````
