# 📅 **Part 1: Date Handling in Pandas**

---

## 🗓️ **Creating Date Ranges**

The `pd.date_range()` function is essential for time series analysis!

---


```python
# =============================================================================
# CREATING DATE RANGES
# =============================================================================
# Purpose: Demonstrate various ways to create date ranges

import pandas as pd

print("=" * 70)
print("CREATING DATE RANGES")
print("=" * 70)
print()

# Basic date range
print("📅 1. Daily date range:")
date_range_daily = pd.date_range(start='2023-01-01', end='2023-01-10', freq='D')
print(date_range_daily)
print(f"   Length: {len(date_range_daily)} days")
print()

# Monthly range
print("📅 2. Monthly date range:")
date_range_monthly = pd.date_range(start='2023-01-01', periods=6, freq='M')
print(date_range_monthly)
print()

# Weekly range
print("📅 3. Weekly date range:")
date_range_weekly = pd.date_range(start='2023-01-01', periods=8, freq='W')
print(date_range_weekly)
print()

# Hourly range
print("⏰ 4. Hourly date range:")
date_range_hourly = pd.date_range(start='2023-01-01 00:00', periods=5, freq='H')
print(date_range_hourly)
print()

# Business days (excludes weekends)
print("💼 5. Business days only:")
date_range_business = pd.date_range(start='2023-01-01', periods=10, freq='B')
print(date_range_business)
print()

print("=" * 70)
```

---

## 🔍 **Extracting Date Components**

The `.dt` accessor unlocks all date/time components!

---


```python
# =============================================================================
# EXTRACTING DATE COMPONENTS
# =============================================================================
# Purpose: Demonstrate extracting day, month, year, weekday, etc.

import pandas as pd

print("=" * 70)
print("EXTRACTING DATE COMPONENTS")
print("=" * 70)
print()

# Create DataFrame with dates
data = {'Date': ['2023-01-15', '2023-02-28', '2023-03-20', '2023-12-25']}
df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])

# Extract components
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['MonthName'] = df['Date'].dt.month_name()
df['DayName'] = df['Date'].dt.day_name()
df['Quarter'] = df['Date'].dt.quarter

print("📅 Date Components:")
print(df)
print()

# Weekday analysis
df_week = pd.DataFrame({'Date': pd.date_range(start='2023-01-01', periods=7)})
df_week['Weekday'] = df_week['Date'].dt.weekday  # Monday=0, Sunday=6
df_week['DayName'] = df_week['Date'].dt.day_name()
df_week['IsWeekend'] = df_week['Date'].dt.weekday >= 5

print("📆 Weekday Analysis:")
print(df_week)
print()

print("💡 Note: Monday=0, Tuesday=1, ..., Sunday=6")
print()

print("=" * 70)
```

# ⏱️ **Part 2: TimeDelta Operations**

---

## 🕐 **Working with Time Intervals**

TimeDelta represents duration - the difference between two times!

---


```python
# =============================================================================
# TIMEDELTA - TIME INTERVALS
# =============================================================================
# Purpose: Demonstrate creating and using TimeDelta

import pandas as pd
import numpy as np

print("=" * 70)
print("TIMEDELTA OPERATIONS")
print("=" * 70)
print()

# Creating TimeDelta
print("⏱️ Creating TimeDelta:")
delta_days = pd.Timedelta(days=3)
delta_hours = pd.Timedelta(hours=12)
delta_mixed = pd.Timedelta(days=2, hours=6, minutes=30)

print(f"3 days: {delta_days}")
print(f"12 hours: {delta_hours}")
print(f"2 days, 6 hours, 30 min: {delta_mixed}")
print()

# Date arithmetic
print("📅 Date Arithmetic:")
df = pd.DataFrame({
    'OrderDate': pd.date_range(start='2023-01-01', periods=5, freq='D')
})

df['ShipDate'] = df['OrderDate'] + pd.Timedelta(days=3)
df['PreviousDay'] = df['OrderDate'] - pd.Timedelta(days=1)
df['DeliveryDate'] = df['OrderDate'] + pd.Timedelta(weeks=1, days=2)

print(df)
print()

# Difference between dates
print("🔢 Calculating Time Differences:")
start_date = pd.to_datetime('2023-01-01')
end_date = pd.to_datetime('2023-01-15')
difference = end_date - start_date

print(f"Start: {start_date.date()}")
print(f"End: {end_date.date()}")
print(f"Difference: {difference} ({difference.days} days)")
print()

print("=" * 70)
```

```python
# =============================================================================
# RESAMPLING TIME SERIES
# =============================================================================
# Purpose: Demonstrate resampling time series data

import pandas as pd
import numpy as np

print("=" * 70)
print("RESAMPLING TIME SERIES DATA")
print("=" * 70)
print()

# Create hourly time series data
dates = pd.date_range(start='2023-01-01', periods=48, freq='H')
df_hourly = pd.DataFrame({
    'DateTime': dates,
    'Sales': np.random.randint(10, 100, 48),
    'Visitors': np.random.randint(50, 200, 48)
})

df_hourly.set_index('DateTime', inplace=True)

print("📊 Original (hourly data - first 10 rows):")
print(df_hourly.head(10))
print()

# Resample to daily
df_daily = df_hourly.resample('D').sum()
print("📅 Resampled to Daily (sum):")
print(df_daily)
print()

# Resample to 6-hour periods
df_6h = df_hourly.resample('6H').mean()
print("⏰ Resampled to 6-Hour Periods (mean):")
print(df_6h)
print()

print("💡 Resampling options: 'H'=hour, 'D'=day, 'W'=week, 'M'=month")
print("💡 Aggregations: sum(), mean(), min(), max(), count()")
print()

print("=" * 70)
```

# 🏷️ **Part 3: Categorical Data Handling**

---

## 📊 **Working with Categories**

Categorical data represents distinct groups or categories!

---


```python
# =============================================================================
# CATEGORICAL DATA HANDLING
# =============================================================================
# Purpose: Demonstrate working with categorical variables

import pandas as pd

print("=" * 70)
print("CATEGORICAL DATA HANDLING")
print("=" * 70)
print()

# Create categorical variable
print("🏷️ 1. Creating Categorical Variable:")
categories = ['Low', 'Medium', 'High']
values = ['Low', 'Medium', 'High', 'Low', 'High', 'Medium', 'Low']
cat_variable = pd.Categorical(values, categories=categories, ordered=True)

print(f"Categories: {cat_variable}")
print(f"Ordered: {cat_variable.ordered}")
print(f"Unique categories: {cat_variable.categories}")
print()

# Value counts
print("📊 2. Counting Categories:")
df = pd.DataFrame({'Product': ['Laptop', 'Phone', 'Laptop', 'Tablet', 'Phone', 'Laptop']})
counts = df['Product'].value_counts()
print(counts)
print()

# Dummy variables (one-hot encoding)
print("🔢 3. Creating Dummy Variables (One-Hot Encoding):")
df_dummies = pd.get_dummies(df['Product'], prefix='Product')
print(df_dummies)
print()

print("💡 Each category becomes a binary (0/1) column")
print("   Essential for machine learning!")
print()

# Label encoding
print("🏷️ 4. Label Encoding:")
df['Product_Code'] = df['Product'].astype('category').cat.codes
print(df)
print()

print("💡 Each category gets a unique number (0, 1, 2, ...)")
print()

print("=" * 70)
```
