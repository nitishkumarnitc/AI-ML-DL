# 📝 **Part 1: Text Data Operations**

---

## 🔤 **The `.str` Accessor - Your Text Toolkit**

All string operations use `.str` accessor!

---


```
# =============================================================================
# TEXT DATA OPERATIONS
# =============================================================================
# Purpose: Demonstrate string operations with .str accessor

import pandas as pd

print("=" * 70)
print("TEXT DATA OPERATIONS")
print("=" * 70)
print()

# Create sample DataFrame
df = pd.DataFrame({
    'Name': ['Alice Smith', 'bob jones', 'CHARLIE BROWN', 'Diana Lee'],
    'Email': ['alice@email.com', 'bob@email.com', 'charlie@email.com', 'diana@email.com'],
    'Product': ['Laptop Pro 15', 'Phone X', 'Tablet Air', 'Monitor 4K']
})

print("📊 Original Data:")
print(df)
print()

# String length
df['Name_Length'] = df['Name'].str.len()
print("📏 String Length:")
print(df[['Name', 'Name_Length']])
print()

# Case conversion
df['Name_Upper'] = df['Name'].str.upper()
df['Name_Lower'] = df['Name'].str.lower()
df['Name_Title'] = df['Name'].str.title()

print("🔤 Case Conversion:")
print(df[['Name', 'Name_Upper', 'Name_Lower', 'Name_Title']].head(2))
print()

# Contains substring
df['Has_Pro'] = df['Product'].str.contains('Pro')
df['Has_Number'] = df['Product'].str.contains(r'\d')  # Regex for digits

print("🔍 Contains Substring:")
print(df[['Product', 'Has_Pro', 'Has_Number']])
print()

# String splitting
df['First_Name'] = df['Name'].str.split().str[0]
df['Last_Name'] = df['Name'].str.split().str[-1]

print("✂️ String Splitting:")
print(df[['Name', 'First_Name', 'Last_Name']])
print()

# Replace text
df['Email_Domain'] = df['Email'].str.replace('@email.com', '@company.com')
print("🔄 String Replace:")
print(df[['Email', 'Email_Domain']])
print()

# Strip whitespace
df_space = pd.DataFrame({'Text': ['  hello  ', 'world  ', '  pandas']})
df_space['Stripped'] = df_space['Text'].str.strip()
print("✨ Strip Whitespace:")
print(df_space)
print()

print("=" * 70)
```

# 🔄 **Part 2: Iteration (Use Wisely!)**

---

## ⚡ **Rule #1: Avoid Loops, Use Vectorization!**

**Vectorized operations are 100x faster than loops!**

---


```
# =============================================================================
# ITERATION - THE RIGHT WAY
# =============================================================================
# Purpose: Show when to use vectorization vs iteration

import pandas as pd
import numpy as np

print("=" * 70)
print("ITERATION IN PANDAS")
print("=" * 70)
print()

# Sample DataFrame
df = pd.DataFrame({
    'Value': [10, 20, 30, 40, 50],
    'Category': ['A', 'B', 'A', 'C', 'B']
})

# ❌ BAD: Using loops (SLOW!)
print("❌ AVOID: Using loops (slow)")
print("   for index, row in df.iterrows():")
print("       # Process each row")
print()

# ✅ GOOD: Vectorized operations (FAST!)
print("✅ PREFER: Vectorized operations (fast)")
df['Doubled'] = df['Value'] * 2
df['PlusHundred'] = df['Value'] + 100
print(df)
print()

# Using .apply() when necessary
print("🔧 Using .apply() for custom functions:")
df['Category_Length'] = df['Category'].apply(lambda x: len(x))
print(df[['Category', 'Category_Length']])
print()

# Vectorized conditional logic
print("✅ Vectorized conditionals (np.where):")
df['Size'] = np.where(df['Value'] > 30, 'Large', 'Small')
print(df[['Value', 'Size']])
print()

print("💡 Key Takeaway: Vectorize whenever possible!")
print("   Only use .apply() or loops when absolutely necessary")
print()

print("=" * 70)
```

# 📊 **Part 3: Sorting Data**

---

## 🔢 **Organizing Your Data**

Sorting is essential for analysis and presentation!

---


```
# =============================================================================
# SORTING DATAFRAMES
# =============================================================================
# Purpose: Demonstrate various sorting methods

import pandas as pd

print("=" * 70)
print("SORTING DATA")
print("=" * 70)
print()

# Create sample DataFrame
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age': [25, 30, 22, 28, 35],
    'Salary': [50000, 60000, 45000, 55000, 70000],
    'Department': ['IT', 'HR', 'IT', 'Finance', 'HR']
})

print("📊 Original DataFrame:")
print(df)
print()

# Sort by single column
print("📈 1. Sort by Age (ascending):")
print(df.sort_values(by='Age'))
print()

print("📉 2. Sort by Salary (descending):")
print(df.sort_values(by='Salary', ascending=False))
print()

# Sort by multiple columns
print("🔢 3. Sort by Department, then by Salary:")
print(df.sort_values(by=['Department', 'Salary'], ascending=[True, False]))
print()

# Sort by index
df_indexed = df.set_index('Name')
print("🏷️ 4. Sort by Index (Name):")
print(df_indexed.sort_index())
print()

# Sort Series
ages = pd.Series([25, 30, 22, 28, 35], index=['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'])
print("📊 5. Sort Series by values:")
print(ages.sort_values(ascending=False))
print()

print("=" * 70)
```

# 📊 **Part 4: Quick Plotting with Pandas**

---

## 📈 **Visualize Your Data Instantly**

Pandas has built-in plotting capabilities!

---


```
# =============================================================================
# PLOTTING WITH PANDAS
# =============================================================================
# Purpose: Demonstrate basic plotting

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Create sample data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [15000, 18000, 22000, 19000, 25000, 28000]
expenses = [12000, 13000, 15000, 14000, 16000, 17000]

df_plot = pd.DataFrame({
    'Month': months,
    'Sales': sales,
    'Expenses': expenses
})

df_plot.set_index('Month', inplace=True)

print("=" * 70)
print("PLOTTING WITH PANDAS")
print("=" * 70)
print()

print("📊 Data to plot:")
print(df_plot)
print()

# Line plot
print("📈 Creating line plot...")
df_plot.plot(kind='line', figsize=(10, 6), title='Monthly Sales & Expenses')
plt.ylabel('Amount ($)')
plt.grid(True)
plt.tight_layout()
plt.show()

# Bar plot
print("📊 Creating bar plot...")
df_plot.plot(kind='bar', figsize=(10, 6), title='Sales vs Expenses')
plt.ylabel('Amount ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print()
print("💡 More plot types: 'scatter', 'hist', 'box', 'area', 'pie'")
print("💡 Use .plot(kind='type') to create different visualizations")
print()

print("=" * 70)
```
