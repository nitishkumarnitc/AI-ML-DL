## 🧮 **Perceptron: The Mathematical Model**

### **How a Perceptron Works:**

```
Input Features (x₁, x₂, ..., xₙ)
        ↓
   Weighted Sum: z = w₁x₁ + w₂x₂ + ... + wₙxₙ + b
        ↓
   Activation Function: ŷ = sign(z)
        ↓
   Output: 0 or 1 (binary classification)
```

### **Key Components:**
1. **Inputs (x)** - Features from your data
2. **Weights (w)** - Learned parameters that determine importance of each feature
3. **Bias (b)** - Shifts the decision boundary
4. **Activation Function** - Converts weighted sum to binary output (0 or 1)

### **Decision Rule:**
```
If (w·x + b) > 0  →  Predict Class 1 (Spam)
If (w·x + b) ≤ 0  →  Predict Class 0 (Not Spam)
```

### **Learning Process:**
The perceptron learns by:
1. Making a prediction
2. Comparing to actual label
3. Adjusting weights when wrong
4. Repeating until convergence

## ⚙️ **Step 4: Data Preprocessing and Splitting**

### **4.1: Check for Missing Values**


````python
# =============================================================================
# CHECK FOR MISSING VALUES
# =============================================================================

print("=" * 70)
print("CHECKING FOR MISSING VALUES")
print("=" * 70)
print()

# Check if any missing values exist in the dataset
has_missing = data.isnull().values.any()

if has_missing:
    print("⚠️  Missing values detected!")
    print(f"   Total missing: {data.isnull().sum().sum()}")
    print()
    print("🔧 Filling missing values with 0...")
    data = data.fillna(0)
    print("✅ Missing values handled!")
else:
    print("✅ No missing values found!")
    print("   Dataset is clean and ready to use.")

print()
````

### **4.2: Separate Features and Target**

We need to split our data into:
- **Features (X):** The 57 input variables used for prediction
- **Target (y):** The output variable we want to predict (spam/not spam)


````python
# =============================================================================
# SEPARATE FEATURES AND TARGET
# =============================================================================

print("=" * 70)
print("SEPARATING FEATURES AND TARGET")
print("=" * 70)
print()

# df_x contains all columns except the last one (features)
df_x = data.iloc[:, :-1]

# df_y contains only the last column (target: spam=1, not spam=0)
df_y = data.iloc[:, -1]

print(f"✅ Data separated successfully!")
print()
print(f"📊 Feature Matrix (X):")
print(f"   - Shape: {df_x.shape}")
print(f"   - Number of features: {df_x.shape[1]}")
print(f"   - Number of samples: {df_x.shape[0]:,}")
print()
print(f"🎯 Target Vector (y):")
print(f"   - Shape: {df_y.shape}")
print(f"   - Unique values: {df_y.unique()}")
print(f"   - Class distribution: {df_y.value_counts().to_dict()}")
print()
````

### **4.3: Train-Test Split**

**Why split the data?**
- **Training Set (60%):** Used to train the perceptron (learn weights)
- **Testing Set (40%):** Used to evaluate performance on unseen data

**Key concept:** The model NEVER sees test data during training!
This simulates real-world deployment where the model encounters new emails.


````python
# =============================================================================
# TRAIN-TEST SPLIT
# =============================================================================

print("=" * 70)
print("SPLITTING DATA INTO TRAIN AND TEST SETS")
print("=" * 70)
print()

# Split: 60% training, 40% testing
# random_state=4 ensures reproducible results
x_train, x_test, y_train, y_test = train_test_split(
    df_x,
    df_y,
    test_size=0.4,      # 40% for testing
    random_state=4       # Seed for reproducibility
)

print(f"✅ Data split successfully!")
print()
print(f"📊 Training Set:")
print(f"   - X_train shape: {x_train.shape}")
print(f"   - y_train shape: {y_train.shape}")
print(f"   - Total training samples: {len(x_train):,} ({len(x_train)/len(data)*100:.1f}%)")
print()
print(f"📊 Testing Set:")
print(f"   - X_test shape: {x_test.shape}")
print(f"   - y_test shape: {y_test.shape}")
print(f"   - Total testing samples: {len(x_test):,} ({len(x_test)/len(data)*100:.1f}%)")
print()
````

### **4.4: Feature Standardization (Scaling)**

**Why standardize?**

The perceptron uses **weighted sums** of features. If features have different scales:
- Large values dominate the calculation
- Small values get ignored
- Model converges slowly or poorly

**Standardization transforms each feature to:**
- **Mean = 0** (centered)
- **Standard deviation = 1** (normalized)

**Formula:** `z = (x - μ) / σ`
- `x` = original value
- `μ` = mean of feature
- `σ` = standard deviation of feature

**⚠️ IMPORTANT:**
- Fit scaler on TRAINING data only
- Apply same transformation to TEST data
- This prevents "data leakage"


````python
# =============================================================================
# FEATURE STANDARDIZATION
# =============================================================================

print("=" * 70)
print("STANDARDIZING FEATURES")
print("=" * 70)
print()

from sklearn.preprocessing import StandardScaler

# Create scaler instance
scaler = StandardScaler()

print("🔧 Fitting scaler on training data...")
# Fit on training data (learns mean and std)
x_train_scaled = scaler.fit_transform(x_train)

print("🔧 Transforming test data with same parameters...")
# Transform test data (uses training mean and std)
x_test_scaled = scaler.transform(x_test)

print()
print("✅ Features standardized successfully!")
print()
print(f"📊 Training Data Statistics (after scaling):")
print(f"   - Mean: ~0.0 (centered)")
print(f"   - Std Dev: ~1.0 (normalized)")
print(f"   - Min: {x_train_scaled.min():.2f}")
print(f"   - Max: {x_train_scaled.max():.2f}")
print()
print(f"💡 Why this matters:")
print(f"   ✅ All features now on same scale")
print(f"   ✅ No single feature dominates")
print(f"   ✅ Faster and more stable convergence")
print(f"   ✅ Better model performance")
print()
````

## 🏗️ **Step 5: Initialize and Train the Perceptron**

Now comes the exciting part - building and training our classifier!

### **What Happens During Training?**

1. **Initialization:** Weights and bias start at random (or zero)
2. **Forward Pass:** For each training sample:
   - Calculate weighted sum: `z = w·x + b`
   - Apply activation function
   - Make prediction
3. **Error Calculation:** Compare prediction to actual label
4. **Weight Update:** Adjust weights to reduce errors
5. **Repeat:** Continue until convergence or max iterations

### **Perceptron Learning Rule:**
```
If prediction is CORRECT:
    → Don't change weights

If prediction is WRONG:
    → Update weights:
      w_new = w_old + η * (y_true - y_pred) * x

Where:
- η (eta) = learning rate (step size)
- y_true = actual label
- y_pred = predicted label
- x = input features
```

### **scikit-learn Perceptron Parameters:**
- `max_iter` - Maximum training iterations (default: 1000)
- `tol` - Tolerance for stopping (default: 1e-3)
- `eta0` - Learning rate (default: 1.0)
- `random_state` - Seed for reproducibility


````python
# =============================================================================
# INITIALIZE AND TRAIN THE PERCEPTRON
# =============================================================================

print("=" * 70)
print("TRAINING THE PERCEPTRON CLASSIFIER")
print("=" * 70)
print()

# Create perceptron instance with default parameters
per = Perceptron(
    max_iter=1000,       # Maximum iterations
    tol=1e-3,            # Tolerance for stopping criterion
    random_state=42      # For reproducible results
)

print("🔧 Perceptron initialized with parameters:")
print(f"   - Max iterations: {per.max_iter}")
print(f"   - Tolerance: {per.tol}")
print(f"   - Random state: {per.random_state}")
print()

print("🚀 Training the model...")
print("   (Learning weights from training data)")
print()

# Train the perceptron on scaled training data
per.fit(x_train_scaled, y_train)

print("✅ Training complete!")
print()
print(f"📊 Learned Parameters:")
print(f"   - Number of features: {len(per.coef_[0])}")
print(f"   - Weight vector shape: {per.coef_.shape}")
print(f"   - Bias term: {per.intercept_[0]:.4f}")
print(f"   - Number of iterations: {per.n_iter_}")
print()
print(f"💡 The perceptron has learned:")
print(f"   ✅ {len(per.coef_[0])} weight values (one per feature)")
print(f"   ✅ 1 bias value (shifts decision boundary)")
print(f"   ✅ Decision boundary to separate spam from non-spam")
print()
````

## 🎯 **Step 6: Make Predictions**

Now that our perceptron is trained, let's use it to classify emails!

### **Prediction Process:**

For each email in the test set:
1. **Input:** 57 feature values (word frequencies, etc.)
2. **Compute:** Weighted sum `z = w·x + b`
3. **Activate:** Apply sign function
4. **Output:** Binary prediction (0 or 1)

### **Why Predict on BOTH Train and Test?**
- **Training predictions:** Check if model learned the training data
- **Testing predictions:** Evaluate generalization to new data
- **Compare both:** Detect overfitting/underfitting


````python
# =============================================================================
# MAKE PREDICTIONS
# =============================================================================

print("=" * 70)
print("MAKING PREDICTIONS")
print("=" * 70)
print()

# Predict on training data
print("🔮 Predicting on training data...")
pred_train = per.predict(x_train_scaled)

# Predict on testing data
print("🔮 Predicting on testing data...")
pred_test = per.predict(x_test_scaled)

print()
print("✅ Predictions complete!")
print()
print(f"📊 Training Predictions:")
print(f"   - Total predictions: {len(pred_train):,}")
print(f"   - Predicted spam: {np.sum(pred_train == 1):,}")
print(f"   - Predicted not spam: {np.sum(pred_train == 0):,}")
print()
print(f"📊 Testing Predictions:")
print(f"   - Total predictions: {len(pred_test):,}")
print(f"   - Predicted spam: {np.sum(pred_test == 1):,}")
print(f"   - Predicted not spam: {np.sum(pred_test == 0):,}")
print()

# Show first 20 predictions vs actual labels
print("🔍 Sample Predictions (First 20 test emails):")
print()
print("   Index | Actual | Predicted | Correct?")
print("   " + "-" * 40)
for i in range(min(20, len(y_test))):
    actual = int(list(y_test)[i])
    predicted = int(pred_test[i])
    correct = "✅" if actual == predicted else "❌"
    actual_label = "Spam" if actual == 1 else "Not Spam"
    pred_label = "Spam" if predicted == 1 else "Not Spam"
    print(f"   {i:3d}   | {actual_label:8s} | {pred_label:9s} | {correct}")
print()
````

### **📊 Understanding the Predictions:**

Each prediction is a **binary value**:
- **0** → Email classified as NOT SPAM (legitimate email)
- **1** → Email classified as SPAM (unwanted email)

The perceptron makes this decision based on:
- The learned weights (importance of each feature)
- The bias term (threshold adjustment)
- The input feature values for each email

**Decision boundary:**
```
If (w·x + b) > 0  →  Predict 1 (SPAM)
If (w·x + b) ≤ 0  →  Predict 0 (NOT SPAM)
```

## 📊 **Step 7: Evaluate Model Performance**

Time to measure how well our perceptron performs!

### **What is Accuracy?**

**Accuracy** = (Number of Correct Predictions) / (Total Predictions)

Formula: `Accuracy = (TP + TN) / (TP + TN + FP + FN)`

Where:
- **TP (True Positives):** Correctly identified spam
- **TN (True Negatives):** Correctly identified non-spam
- **FP (False Positives):** Non-spam incorrectly marked as spam
- **FN (False Negatives):** Spam incorrectly marked as non-spam

### **Why Measure Both Train and Test Accuracy?**

| Scenario | Train Acc | Test Acc | Diagnosis |
|----------|-----------|----------|-----------|
| Good Fit | ~90% | ~90% | ✅ Model generalizes well |
| Overfitting | 99% | 70% | ⚠️ Memorized training data |
| Underfitting | 60% | 60% | ⚠️ Model too simple |


````python
# =============================================================================
# CALCULATE ACCURACY SCORES
# =============================================================================

print("=" * 70)
print("EVALUATING MODEL PERFORMANCE")
print("=" * 70)
print()

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Calculate accuracy scores
accuracy_train = accuracy_score(y_train, pred_train)
accuracy_test = accuracy_score(y_test, pred_test)

print(f"📊 ACCURACY SCORES:")
print(f"   {'='*50}")
print(f"   Training Accuracy:   {accuracy_train:.4f} ({accuracy_train*100:.2f}%)")
print(f"   Testing Accuracy:    {accuracy_test:.4f} ({accuracy_test*100:.2f}%)")
print(f"   {'='*50}")
print()

# Calculate number of correct/incorrect predictions
train_correct = int(accuracy_train * len(y_train))
train_incorrect = len(y_train) - train_correct
test_correct = int(accuracy_test * len(y_test))
test_incorrect = len(y_test) - test_correct

print(f"📈 DETAILED RESULTS:")
print()
print(f"   Training Set:")
print(f"      ✅ Correct predictions:   {train_correct:,} / {len(y_train):,}")
print(f"      ❌ Incorrect predictions: {train_incorrect:,} / {len(y_train):,}")
print()
print(f"   Testing Set:")
print(f"      ✅ Correct predictions:   {test_correct:,} / {len(y_test):,}")
print(f"      ❌ Incorrect predictions: {test_incorrect:,} / {len(y_test):,}")
print()

# Interpretation
print(f"💡 INTERPRETATION:")
print()
diff = abs(accuracy_train - accuracy_test)
if diff < 0.05:
    print(f"   ✅ EXCELLENT! Model generalizes well.")
    print(f"   → Train and test accuracy are similar ({diff*100:.1f}% difference)")
    print(f"   → No signs of overfitting or underfitting")
elif accuracy_train > accuracy_test + 0.1:
    print(f"   ⚠️  Possible overfitting detected!")
    print(f"   → Training accuracy is much higher than test")
    print(f"   → Model may have memorized training data")
else:
    print(f"   ℹ️  Model performance:")
    print(f"   → {diff*100:.1f}% difference between train and test")
    print(f"   → Acceptable for linear classifier")
print()
````

````python
# =============================================================================
# CONFUSION MATRIX ANALYSIS
# =============================================================================

print("=" * 70)
print("CONFUSION MATRIX - DETAILED BREAKDOWN")
print("=" * 70)
print()

# Generate confusion matrix for test set
cm = confusion_matrix(y_test, pred_test)

print("📊 Confusion Matrix:")
print()
print("                 Predicted")
print("               Not Spam  Spam")
print("   Actual    ┌─────────┬─────────┐")
print(f"   Not Spam │  {cm[0][0]:6d}  │  {cm[0][1]:6d}  │  (True Negatives | False Positives)")
print("            ├─────────┼─────────┤")
print(f"   Spam     │  {cm[1][0]:6d}  │  {cm[1][1]:6d}  │  (False Negatives | True Positives)")
print("            └─────────┴─────────┘")
print()

# Calculate metrics from confusion matrix
TN, FP, FN, TP = cm.ravel()

print(f"📈 Breakdown:")
print(f"   True Negatives (TN):  {TN:,} - Correctly identified non-spam")
print(f"   True Positives (TP):  {TP:,} - Correctly identified spam")
print(f"   False Positives (FP): {FP:,} - Non-spam marked as spam (Type I error)")
print(f"   False Negatives (FN): {FN:,} - Spam marked as non-spam (Type II error)")
print()

# Additional metrics
precision = TP / (TP + FP) if (TP + FP) > 0 else 0
recall = TP / (TP + FN) if (TP + FN) > 0 else 0
f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

print(f"🎯 Advanced Metrics:")
print(f"   Precision: {precision:.4f} ({precision*100:.2f}%) - Of predicted spam, how many were actually spam?")
print(f"   Recall:    {recall:.4f} ({recall*100:.2f}%) - Of actual spam, how many did we catch?")
print(f"   F1-Score:  {f1_score:.4f} ({f1_score*100:.2f}%) - Harmonic mean of precision and recall")
print()
````

````python
# =============================================================================
# CLASSIFICATION REPORT
# =============================================================================

print("=" * 70)
print("DETAILED CLASSIFICATION REPORT")
print("=" * 70)
print()

# Generate comprehensive classification report
report = classification_report(y_test, pred_test, target_names=['Not Spam', 'Spam'])
print(report)
print()

print("📚 Understanding the Report:")
print()
print("   • Precision: When model predicts a class, how often is it correct?")
print("   • Recall: Of all actual instances of a class, how many did we find?")
print("   • F1-Score: Balance between precision and recall")
print("   • Support: Number of actual occurrences of each class in test set")
print()
````

### **🔍 Perceptron Limitations:**

⚠️ **What perceptrons CAN'T do:**
- Solve non-linearly separable problems (e.g., XOR problem)
- Capture complex patterns in data
- Learn hierarchical representations

💡 **The solution?**
→ **Multi-layer Neural Networks** (coming in next lessons!)
→ Add hidden layers between input and output
→ Use non-linear activation functions
→ Enable learning of complex patterns
